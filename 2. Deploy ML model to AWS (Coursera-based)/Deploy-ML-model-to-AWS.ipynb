## 6. Deploy ML model (Flask API) into AWS virtual servers.

Before deploying the application to AWS, we need to create a file of requirements (dependancies) for our project - to "tell" AWS Elastic Beanstalk to import all modelues and libraries used in the project, such as: Flask, joblib, etc. It can be done by running the command in the application environment:
- pip3 freeze > requirements.txt

When we will deploy the file "requirements.txt" along with code base into AWS, it informs the server to install all those dependancies onto its cloud service. So, technically, AWS will install all the dependancies automatically rather than the user deployning them manually. That's an awesome time-saver feature!

As a result, for the deployment on AWS, we need 4 files:
1. application.py
2. requirements.txt
3. spam_ham_model.pkl
4. vectorizer.pkl

Those 4 files need to be archived into "version_1.zip" file.

When ready, go to the AWS Management Console >> Services >> Elastic Beanstalk >> select the environment >> click "Uplod and Deploy" >> "Choose file", and proceed with deployment. Again, it might take several minutes to complete.
